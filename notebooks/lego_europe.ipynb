```python
# Import required packages
import openeo
from openeo.processes import process
from IPython.display import Image

# Connect to the back-end
connection = openeo.connect("http://0.0.0.0:8081")
# ToDo: Here you need to authenticate with authenticate_basic() or authenticate_oidc()
connection.authenticate_oidc()

# zoom level
zoom = 9

# European sptial
spatial_extent_east = 15.0
spatial_extent_west = -5.0
spatial_extent_north = 60.0
spatial_extent_south = 35.0
```

```python
# Tiles extractions
# We will simulate the webmercator XYZ calls to get the tiles of the map at the given zoom level
# for the spatial extent

# The following function will calculate the tile numbers for the given zoom level
def get_tile_numbers(zoom, spatial_extent_east, spatial_extent_west, spatial_extent_north, spatial_extent_south):
    # Calculate the tile numbers
    tile_numbers = []
    for x in range(int((spatial_extent_west + 180) / 360 * (2 ** zoom)),
                   int((spatial_extent_east + 180) / 360 * (2 ** zoom))):
        for y in range(int((1 - (spatial_extent_north + 90) / 180) / 2 * (2 ** zoom)),
                       int((1 - (spatial_extent_south + 90) / 180) / 2 * (2 ** zoom))):
            tile_numbers.append((x, y))
    return tile_numbers
# Get the tile numbers
tile_numbers = get_tile_numbers(zoom, spatial_extent_east, spatial_extent_west, spatial_extent_north, spatial_extent_south)
# Print the tile numbers
print("Tile numbers for zoom level {}:".format(zoom))
for tile_number in tile_numbers:
    print("Tile number: {}".format(tile_number))
    # Get the tile URL
    tile_url = "http://localhost:8080/tile/{}/{}/{}.png".format(zoom, tile_number[0], tile_number[1])
```

```python
# Let's define a function that, for a given tile (XYZ), will return the set of subtitles tiles at a given zoom level
# necessary to make a mosaic and form the complete tile
def get_tiles(x, y, z, zoom):
    # Calculate the tile numbers
    tile_numbers = []
    for i in range(0, 2 ** (zoom - z)):
        for j in range(0, 2 ** (zoom - z)):
            tile_numbers.append((x * 2 ** (zoom - z) + i, y * 2 ** (zoom - z) + j))
    return tile_numbers
```
